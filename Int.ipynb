# class Int

```python
############################################################
# Int.py 
# Implements Int object
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################


###########################################################
#  class  Int
###########################################################   
class Int():
    def __init__(self,n:'int' = 0):  
        #ONLY DATA STRUCTURE ALLOWED
        #self._positive
        #self._a
        self._positive = True
        if(n<0):
            self._positive = False
        self._a = self.build(n)    
        

    #############################
    # WRITE All public functions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    #############################

    ##############################################################
    # WRITE All private functtions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    ##############################################################

    def build(self,n:'int')->'list of int':
        if(n<0):
            n = -n
        l=[]
        if(n<10):
            l.append(n)
        else:
            while(n!=0):
                l.append(n%10)
                n = n//10   
        self._reverse(l)
        return self.l
    
    
    def _reverse(self,l)->'list':
        l = l[::-1]
        self.l = l
        return self.l
            
    def int(self)->'integer':
        v=0
        for e in self._a:
            v = 10*v+e
        if(v==0 or self._positive):
            return v 
        return -v
    
    def __getitem__(self,pos)->'int':
        n=len(self)
        assert(pos>=0 and pos<n)
        return self._a[pos]
    
    def __setitem__(self,pos,v)->'None':
        n = len(self)
        assert(pos>=0 and pos<n)
        self._a[pos]= v
        
    def __len__(self)->'int':
        return len(self._a)
    
    def __str__(self)->'String':
        n=self.int()
        sign = "-"
        if(self._positive):
            sign = "+"
        
        s = str(n) + " " + sign + "["
        d = len(self)
        for i in range(d):
            x = self[i]
            s = s + str(x)
            if(i<d-1):
                s = s + "," + " "
                
        s = s + "]"
        return s
   
    def __add__(self,b:'Int')->'Int':
        va = self.int()
        vb = b.int()
        vc = va + vb 
        c = Int(vc)
        return c
    
    def __sub__(self,b:'Int')->'Int':
        va = self.int()
        vb = b.int()
        vc = va - vb 
        c = Int(vc)
        return c
    
    def __lt__(self,b:'Int')->'bool':
        va = self.int()
        vb = b.int()
        return(va<vb)
    
    def __eq__(self,b:'Int')->'bool':
        va = self.int()
        vb = b.int()
        return(va==vb)
    
        
            

              
```
